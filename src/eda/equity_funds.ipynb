```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
df = pd.read_csv("../data/canadian_mutual_funds_price_history.csv")
df
```

```python
fund_tickers = df['ticker'].unique().tolist()
print(fund_tickers)


len(fund_tickers)


```

```python
yf.Ticker(fund_tickers[0]).info
```

```python
# plot the trend of one fund
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
fund_df = df[df['ticker'] == fund_tickers[0]]
plt.plot(pd.to_datetime(fund_df["Date"]), fund_df["Close"])
plt.title(f"Price Trend of {fund_tickers[0]}")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid()
plt.show()  




```

```python
# use ols from statsmodels to analyze trend
import statsmodels.api as sm

X = pd.to_datetime(fund_df["Date"]).map(pd.Timestamp.timestamp)
y = fund_df["Close"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())  


# plot the regression line
plt.figure(figsize=(12,6))
plt.plot(pd.to_datetime(fund_df["Date"]), fund_df["Close"], label="Actual Prices")
plt.plot(pd.to_datetime(fund_df["Date"]), model.predict(X), color='red', label="Trend Line")
plt.title(f"Price Trend of {fund_tickers[0]} with Trend Line")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.grid()
plt.show()
```

```python
# add moving average 50 days and 200 days
fund_df['MA50'] = fund_df['Close'].rolling(window=50).mean()
fund_df['MA200'] = fund_df['Close'].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(pd.to_datetime(fund_df["Date"]), fund_df["Close"], label="Actual Prices")
plt.plot(pd.to_datetime(fund_df["Date"]), fund_df['MA50'], label        ="50-Day MA")
plt.plot(pd.to_datetime(fund_df["Date"]), fund_df['MA200'], label       ="200-Day MA")
plt.title(f"Price Trend of {fund_tickers[0]} with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.grid()
plt.show()
```

```python
returns = df["Close"].pct_change().dropna()

# Estimate parameters
mu = returns.mean()
sigma = returns.std()

S0 = df["Close"].iloc[-1]  # Last known price
days = 30  # Number of days to simulate (e.g., 1 year)
simulations = 50  # Number of simulation paths

simulated_paths = np.zeros((simulations, days))
for i in range(simulations):
    prices = [S0]
    for _ in range(days):
        next_price = prices[-1] * np.exp(np.random.normal(mu, sigma))
        prices.append(next_price)
    simulated_paths[i] = prices[1:]

# Plot simulated price paths
plt.figure(figsize=(10, 6))
for i in range(simulations):
    plt.plot(simulated_paths[i], color="blue", alpha=0.3)
plt.title("Monte Carlo Simulated Price Paths")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "AAPL"
yf_ticker = yf.Ticker(ticker)

# Get historical close prices
hist = yf_ticker.history(period="6mo", interval="1d")

# Get option chain for a specific expiry
expiry = yf_ticker.options[0]
opt_chain = yf_ticker.option_chain(expiry)
calls = opt_chain.calls

# Example: plot close price and call open interest for a strike
strike = 150
call_strike = calls[calls['strike'] == strike]

plt.figure(figsize=(12,6))
plt.plot(hist.index, hist['Close'], label="Close Price")
plt.scatter([expiry], [call_strike['openInterest'].values[0]], color='red', label="OI at Expiry")
plt.title(f"{ticker} Close Price & Call Open Interest (Strike {strike})")
plt.xlabel("Date")
plt.ylabel("Price / Open Interest")
plt.legend()
plt.show()
```
